```python
import numpy as np

import matplotlib.pyplot as plt

# Parameters
v = 0.1  # velocity
dt = 0.01  # time step
k = 10  # target steps from origin
num_simulations = 10000  # number of simulations

# Probabilities
p_plus = v * dt
p_zero = 1 - v * dt

# Function to simulate one walk
def simulate_walk(k, p_plus, p_zero):
    position = 0
    steps = 0
    while position < k:
        if np.random.rand() < p_plus:
            position += 1
        steps += 1
    return steps

# Run simulations
results = [simulate_walk(k, p_plus, p_zero) for _ in range(num_simulations)]

# Plot histogram
plt.hist(results, bins=30, edgecolor='black')
plt.xlabel('Number of Steps')
plt.ylabel('Frequency')
plt.title('Histogram of Steps to Reach {} Steps from Origin'.format(k))
plt.show()
```
